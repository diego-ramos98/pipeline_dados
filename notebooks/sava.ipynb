```python
import mysql.connector
```

```python
cnx = mysql.connector.connect(
        host = "localhost",
        user = "root",
        password = "123456"
    )
print(cnx)
```

```python
cursor = cnx.cursor()
```

```python
cursor.execute(f"CREATE DATABASE IF NOT EXISTS db_produtos_varejo;")
```

```python
cursor.execute("SHOW DATABASES")
list_database = [] 
for x in cursor:
    list_database.append(x)

list_database
```

```python
cursor.execute(
        """
            CREATE TABLE  IF NOT EXISTS db_produtos_varejo.tb_produtos(
	                ID int,
                    NOME_PRODUTO varchar(100),
                    CATEGORIA_PRODUTO varchar(60),
                    PRECO_PRODUTO float,
                    QUANTIDADE_ESTOQUE int,
                    FILIAL varchar(20),

                    primary key (id)
            )

        """
    )
```

```python
cursor.execute("USE db_produtos_varejo")
cursor.execute("SHOW TABLES")

list_tables = []

for x in cursor:
   list_tables.append(x)

list_tables
```
